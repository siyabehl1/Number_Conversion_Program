```python
#Siya Behl (author of following program)
#Project 1: Number Converter
#CMSC 206 40397
#This is my Number conversion project which takes user input and converts it to the user's desired base

def decimal_to_base(decimal, indicated_base):
 #Detect variable holds values that the code is going to use to form the Hex value
  detect = "0123456789ABCDEF"
  if decimal == 0:
    return "0"
  result = ""
  while decimal > 0:
    #Calculates remainer when decimal number needs to be divided by base using modulus
    remainder = decimal % indicated_base
    #Adds the corresponding digit from the digits string to the beginning of the result string
    result = detect[remainder] + result
    #Does integer division to update the decimal number
    decimal = decimal // indicated_base
  return result

def base_to_decimal(number, indicated_base):
  # Converts a number in the specified base to decimal
  decimal = 0
  power = 0
  #Based on the selected if/elif the following operation will be performed on the input
  #Iterates over each digit in the given number, starting with the one furthest on the right
  for digit in reversed(number):
    #Ensures that the correct exponent is applied to the input value
    #Calculates the decimal value by multiplying it with the power of the base and adding it to the running total decimal
    decimal += int(digit) * (indicated_base ** power)
    #Increases the power by 1 before progressing to the next number
    power += 1
  return decimal

#Prints the menu options and prompts for user input
def my_number_converter():
  # Function to perform number conversions
  print("***** Siya's Number Converter *****")
  print("")
  print("1. Decimal to Hexadecimal")
  print("2. Hexadecimal to Decimal")
  print("3. Decimal to Octal")
  print("4. Octal to Decimal")
  print("5. Decimal to Binary")
  print("6. Binary to Decimal")
  #EXTRA CREDIT FEATURE: Allows user to convert from binary to decimal
  print("7. Octal to Binary")
  print("0. Exit")
  print("")
  print("***********************************")

  #Asks user to indicate which conversion they want to perform
  choice = input("Enter your conversion choice: ")

  #Code to exit the program if 0 is selected from menu
  if choice == "0":
    print("Exiting ...")
    return

  #Checks menu choice to make sure it is in the rage and calls for valid input if not given
  if choice not in ["1", "2", "3", "4", "5", "6", "7"]:
    print("Invalid choice. Please try again.\n")
    my_number_converter()

  #Asks the user for a number to convert
  number = input("Enter your number choice: ")

  if not number.isdigit() and not all(char in "01234567ABCDEF" for char in number):
    print("Invalid number format. Please try again.\n")
    my_number_converter()

  #Makes sure to store the number value as an integer
  decimal_num = int(number)

  #Ensures that the number given does not exceede the range of 0-2048
  if decimal_num < 0 or decimal_num > 2048:
    print("Number out of range. Please try again.\n")
    my_number_converter()

  #Series of if-else statments that ensure that the correct function is being executed on the input according to selected conversion
  #Each choice corresponds to the according value in the Conversion Menu options
  if choice == "1":
    result = decimal_to_base(decimal_num, 16).upper()
  elif choice == "2":
    result = str(base_to_decimal(number, 16))
  elif choice == "3":
    result = decimal_to_base(decimal_num, 8)
  elif choice == "4":
    result = str(base_to_decimal(number, 8))
  elif choice == "5":
    result = decimal_to_base(decimal_num, 2)
  elif choice == "6":
    result = str(base_to_decimal(number, 2))
  elif choice == "7":
    result = decimal_to_base(base_to_decimal(number, 8), 2)

  #Prints result of the conversion calculation
  print("Conversion Result:", result)
  print()

  #EXTRA CREDIT FEATURE: Asks the user to perform another conversion
  another_conversion = input("Do you want to perform another conversion? (Y/N): ")
  #Makes sure that 'y' and 'Y' will recall the converter sequence
  if another_conversion.upper() == "Y":
    print()
    my_number_converter()
  else:
    print("Exiting.....")

#Function compiles number converter
my_number_converter()
```
